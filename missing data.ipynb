```python
df = pd.read_csv("weather_missing.csv", parse_dates = ["day"])
df
```

```python
type(df.day[0])
```

```python
df.set_index('day', inplace = True)    #setting day as index
```

```python
df
```

```python
new_df = df.fillna(0)             # replace nan with 0
new_df
```

```python
new_df = df.fillna ({
    'Temperature': 0,
    'WindSpeedMPH' : 0,
    'Events' : 'no events'  
})
new_df
```

```python
new_df = df.fillna(method = 'ffill')              # replace nan with carry forward previous day value or "pad"
new_df
```

```python
new_df = df.fillna(method = 'backfill')              # replace nan with carry forward next day value or "bfill"
new_df
```

```python
values = {'Temperature: 1', 'WindSpeedMPH': 2, 'Events': 3}
new_df = df.fillna(value = values)
new_df
```

```python
new_df = df.fillna(method = 'backfill', axis = "columns")     # replace nan with horizontal value or "bfill"
new_df
```

```python
new_df = df.fillna(method = 'ffill', axis = "columns")     # replace nan with horizontal value or "ffill"
new_df
```

```python
new_df = df.fillna(method = 'ffill', limit =1)     # replace nan upto on repeated value
new_df
```

```python
new_df = df.fillna(method = 'ffill', limit =2)     # replace nan upto 2repeated value
new_df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df1 = df.interpolate(method="time")                                #error
new_df1
```

```python
new_df = df.dropna()                   #show only valid data
new_df
```

```python
new_df = df.dropna(how = "all")                  # remove data having all nan
new_df
```

```python
new_df = df.dropna(thresh = 1)                   #keep atleast one valid value
new_df
```

```python
new_df = df.dropna(thresh = 2)                 # need two valid values
new_df
```
