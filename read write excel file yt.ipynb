```python
import pandas as pd
def convert_people_cell(cell):
    if cell =="n.a.":
        return 'sam walton'
    return cell
def convert_eps_cell(cell):
    if cell == "not availale":
        return None
    return cell
df=pd.read_excel("stock_excel.xlsx", "Sheet1", converters = {'people': convert_people_cell, 'eps': convert_eps_cell})
df
```

```python
df.to_excel("stock_excel_new.xlsx", sheet_name = "stocks", startrow = 1,startcol = 2, index = False)
```

```python
#write to different sheets in same excel file
import pandas as pd
df_stocks = pd.DataFrame({"eps":[27,25,25.5], 
                          "pe": [30.25,50.25,48.25],
                          "price" : [84,98,78], 
                          "tickers": ["google", "wmt","msft"]})

df_weather = pd.DataFrame({"day":['1/7/2017','1/2/2017', '1/3/2017'], 
                           "event": ["rain", "snow", "sunny"],
                           "temperature": [32,25,34]})
```

```python
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name = "stocks")
    df_weather.to_excel(writer, sheet_name = "weather")
```
